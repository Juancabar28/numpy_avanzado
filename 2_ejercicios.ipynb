### Utiliza Python y Numpy para realizar las siguientes operaciones:
1. Dado el arreglo `a = np.array([1, 2, 3, 4])`, determina si ninguno de los elementos es cero.
2. Dado el arreglo `a = np.array([0, 0, 2, 0])`, determina si contiene algún elemento que no sea cero.
3. Analiza el arreglo `a = np.array([1, 2, np.nan, np.inf])` y determina si contiene valores que sean infinitos o valores no numéricos (NaN).
4. Dado el arreglo `a = np.array([1, 2, np.inf])`, determina si contiene valores infinitos.
5. Dado el arreglo `a = np.array([1, np.nan, 3])`, determina si contiene valores NaN.
6. Dados los arreglos `a = np.array([1, 3, 5])` y `b = np.array([2, 3, 4])`, realiza una comparación elemento a elemento contemplando mayor, mayor o igual, menor y menor o igual.
7. Crea un arreglo que contenga los valores 1, 7, 13 y 105, y determina la cantidad de memoria que ocupa el arreglo.
8. Crea un arreglo de 10 ceros; uno de 10 unos; y uno de 10 cincos.
9. Crea un arreglo que contenga todos los números enteros entre 30 y 70.
10. Crea un arreglo que contenga los números pares entre 30 y 70.
11. Crea una matriz identidad de 3x3.
12. Genera un número aleatorio entre 0 y 1.
13. Crea un arreglo de 15 números aleatorios tomados de una distribución normal estándar.
14. Crea un arreglo con valores entre 15 y 55, e imprime todos los valores excepto el primero y el último.

16. Crea un vector con valores de 0 a 20, y cambia el signo de los números de 9 a 15.

20. Dado el arreglo `m = np.array([[1, 2, 3], [4, 5, 6]])`, encuentra el número de filas y columnas de la matriz.

```python
import numpy as np
```

```python
# ejercicio 1
a = np.array([1, 2, 3, 4])
comparacion = (a != 0)

resultado = np.all(comparacion)

print(f"Arreglo: {a}")
print(f"Comparación (a != 0): {comparacion}")
print(f"Resultado (np.all(comparacion)): {resultado}")
```

```python
# ejercicio 2
a = np.array([1, 2, 3, 4])
comparacion = (a != 0)

resultado = np.any(comparacion)

print(f"Arreglo: {a}")
print(f"Comparación (a != 0): {comparacion}")
print(f"Resultado (np.all(comparacion)): {resultado}")
```

```python
# ejercicio 3
a = np.array([1, 2, np.nan, np.inf])
infinitos_check = np.isinf(a)
hay_infinitos = np.any(infinitos_check)

nan_check = np.isnan(a)
hay_nan = np.any(nan_check)

combinado_check = np.any(infinitos_check | nan_check)

print(f"Arreglo: {a}")
print(f"¿Contiene infinitos? (np.isinf): {infinitos_check}")
print(f"Resultado 'Hay Infinitos': {hay_infinitos}")
print("-" * 30)
print(f"¿Contiene NaN? (np.isnan): {nan_check}")
print(f"Resultado 'Hay NaN': {hay_nan}")
print("-" * 30)
print(f"¿Contiene valores infinitos O no numéricos?: {combinado_check}")
```

```python
# ejercicio 4
a = np.array([1, 2, np.inf])

es_infinito = np.isinf(a)

contiene_infinito = np.any(es_infinito)

print(f"Arreglo: {a}")
print(f"Verificación elemento a elemento (np.isinf): {es_infinito}")
print(f"Resultado final (np.any): {contiene_infinito}")
```

```python
# ejercicio 5
a = np.array([1, np.nan, 3])

es_nan = np.isnan(a)

contiene_nan = np.any(es_nan)

print(f"Arreglo: {a}")
print(f"Verificación elemento a elemento (np.isnan): {es_nan}")
print(f"Resultado final (np.any): {contiene_nan}")
```

```python
# ejercicio 6
a = np.array([1, 3, 5])
b = np.array([2, 3, 4])

mayor_que = a > b
mayor_o_igual_que = a >= b
menor_que = a < b
menor_o_igual_que = a <= b

print(f"Arreglo a: {a}")
print(f"Arreglo b: {b}")
print("-" * 40)
print(f"a > b (Mayor que):         {mayor_que}")
print(f"a >= b (Mayor o igual que):{mayor_o_igual_que}")
print(f"a < b (Menor que):         {menor_que}")
print(f"a <= b (Menor o igual que):{menor_o_igual_que}")
```

```python
# ejercicio 7
a = np.array([1, 7, 13, 105])

num_elementos = a.size
bytes_por_elemento = a.itemsize
memoria_total = a.nbytes

print(f"Arreglo: {a}")
print(f"Tipo de datos de los elementos: {a.dtype}")
print("-" * 40)
print(f"Número de elementos (a.size): {num_elementos}")
print(f"Bytes por elemento (a.itemsize): {bytes_por_elemento} bytes")
print(f"Bytes totales ocupados (a.nbytes): {memoria_total} bytes")

# Verificación manual
print(f"Verificación manual (size * itemsize): {num_elementos * bytes_por_elemento} bytes")
```

```python
# ejercicio 8
arreglos_de_ceros = np.zeros(10)

arreglos_de_unos = np.ones(10)

arreglos_de_cincos = np.full(10, 5)

print("1. Arreglo de 10 Ceros:")
print(arreglos_de_ceros)
print("-" * 30)
print("2. Arreglo de 10 Unos:")
print(arreglos_de_unos)
print("-" * 30)
print("3. Arreglo de 10 Cincos:")
print(arreglos_de_cincos)
```

```python
# ejercicio 9
arreglo_secuencial = np.arange(30, 71)

print(f"Arreglo creado (desde 30 hasta 70):")
print(arreglo_secuencial)
print("-" * 40)
print(f"Primer elemento: {arreglo_secuencial[0]}")
print(f"Último elemento: {arreglo_secuencial[-1]}")
```

```python
# ejercicio 11
matriz_identidad = np.eye(3)

print("Matriz Identidad 3x3:")
print(matriz_identidad)
```

```python
# ejercicio 12
numero_aleatorio = np.random.rand()

print(f"Número aleatorio entre 0 y 1: {numero_aleatorio}")
```

```python
# ejercicio 13
arreglo_normal = np.random.randn(15)

print(f"Arreglo de 15 números con distribución Normal Estándar:")
print(arreglo_normal)
print("-" * 60)
print(f"Número de elementos: {arreglo_normal.size}")
print(f"Media aproximada: {np.mean(arreglo_normal):.4f}")
print(f"Desviación estándar aproximada: {np.std(arreglo_normal):.4f}")
```

```python
# ejercicio 14
a = np.arange(15, 56)

valores_centrales = a[1:-1]

print(f"Arreglo completo (15 a 55):")
print(a)
print("-" * 60)
print(f"Valores excluyendo el primero y el último:")
print(valores_centrales)
print("-" * 60)
print(f"Primer valor original (excluido): {a[0]}")
print(f"Último valor original (excluido): {a[-1]}")
```

```python
# ejercicio 15
vector_equiespaciado = np.linspace(5, 50, 10)

print(f"Vector equitativamente distribuido (Longitud: 10):")
print(vector_equiespaciado)
print("-" * 60)
print(f"Primer elemento: {vector_equiespaciado[0]}")
print(f"Último elemento: {vector_equiespaciado[-1]}")
print(f"Número de elementos: {vector_equiespaciado.size}")
```

```python
# ejercicio 16
vector = np.arange(21)

print(f"Vector original (0 a 20):")
print(vector)
print("-" * 60)

vector[9:16] = vector[9:16] * -1

print(f"Vector modificado (Valores de 9 a 15 con signo cambiado):")
print(vector)
```

```python
# ejercicio 19
m = np.array([[1, 2],
              [3, 4]])

matriz_par_check = (m % 2) == 0

print(f"Matriz original:")
print(m)
print("-" * 30)
print(f"Resultado (Matriz Booleana de Pares):")
print(matriz_par_check)
```

```python
# ejercicio 20
m = np.array([[1, 2, 3],
              [4, 5, 6]])

dimensiones = m.shape

num_filas = dimensiones[0]
num_columnas = dimensiones[1]

print(f"Matriz m:")
print(m)
print("-" * 30)
print(f"Dimensiones de la matriz (m.shape): {dimensiones}")
print(f"Número de filas: {num_filas}")
print(f"Número de columnas: {num_columnas}")
```
